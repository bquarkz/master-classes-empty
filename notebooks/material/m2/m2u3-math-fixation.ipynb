### 2. Proportional Stratified Allocation

A population is divided into three strata:

- Stratum 1: $N_1 = 500$
- Stratum 2: $N_2 = 1{,}000$
- Stratum 3: $N_3 = 1{,}500$

You want a total sample of $n = 200$ using **proportional stratified sampling**.

1. Compute $N = N_1 + N_2 + N_3$.
2. Compute the stratum sample sizes:
   $$
   n_h = \frac{N_h}{N}\,n
   $$
   for $h = 1,2,3$.
3. Check that $n_1 + n_2 + n_3 = 200$ (up to rounding).

1.
$$
N = N_1 + N_2 + N_3 = 500 + 1000 + 1500 = 3000
$$

2.
$$
n_1 = \frac{N_1}{N}\,n = \frac{500}{3000}\,200 = 0.17*200 = 33.34
$$
$$
n_2 = \frac{N_2}{N}\,n = \frac{1000}{3000}\,200 = \frac{200}{3} = 66.67
$$
$$
n_3 = \frac{N_3}{N}\,n = \frac{1500}{3000}\,200 = \frac{200}{2} = 100
$$

3.
$$
n_1 + n_2 + n_3 = 33.34 + 66.67 + 100 = 200.01 = n
$$

### 3. Homogeneous Strata

Explain why, in stratified sampling, it is desirable that:

- Units **within** each stratum are similar to each other (homogeneous), but
- Strata are **different from each other**.

Give a simple example of a good stratification variable (e.g. age group, income band, region) and explain your choice quickly.

> Because the variance inside each stratum is small, and it gives us a more stable sample when sampling each stratum in proportion to its population (when keeping the proportion of the population in the sample).

> As an example, we could use an age group for a population of 1000 runners. When sampling that population, we could use the age group as a stratification variable and keep the proportion of the population in our sample.

> Let me go for a more complete explanation. Into the population of 1000 runners we have 200 the age of the 10s, 500 the age of the 20s and the remain of 300 the age of the 30s. If you SRS (20 units) the whole population without stratification, you could get something like: 12 the age of the 10s, 0 the age of the 20s and 8 the age of the 30s; It is not a good sample since there are no units from the 20s. It is SRS that is ok, but it will be great to have a better (stable, more representative) sample. That is why stratification came handy, we could divide the population by age: 200, 500 and 300, and keep those proportions when sampling (0.2, 0.5 and 0.3, respectively). So the new sampling (from each stratum) could be something like: the 10s ($ 0.2 * 20 = 4 $), the 20s ($ 0.5 * 20 = 10 $) and the 30s ($ 0.3 * 20 = 6 $); what will give us a much more stable sampling.

### 4. Cluster vs Stratified Sampling

In your own words, describe the main difference between **stratified sampling** and **cluster sampling** regarding:

- How groups are formed
- How we sample from those groups (within or entire groups)

Give an example for each where it would be more natural to use that design.

> For Stratified Sampling we try to reduce the variance for each stratum at maximum (keep it as small as possible). What is not really the case for cluster, where division is taken by some factor that is not part of the statistical approach itself (logistical, political and economic reasons, for example), which does not help to reduce the variance. For Stratified Sampling we sample for each stratum. For Clusters, we sample the entire group.

> Using the same example of runners, the Stratified Sampling example is above (and we could say that is for a single city). We could add the Cluster example saying that our population is the whole country, and it will be impossible to let people go to each city and sample them individually (that is not logistically or economically viable). In that way we use the cities as a cluster, and sample cities (SRS) from the country. And for each city in the sample, we run SRS (20 units) without (one-stage) or with (two-stage) stratification.

> In general, if we compare both methods Stratified Sampling and Cluster against a SRS for the whole population, we could say:
>  * Stratified Sampling: could reduce variance against SRS, which improves precision (and better samplings).
>  * Cluster sampling: often increases variance if compared with SRS, but it is more cost/fieldwork effective (and sometimes the only way to get a sample).

> In short, for a valid inference and representative sampling (possibility to infer propositions from samples to the whole population), we could generate clusters, SRS some of them, and further run Stratified Sampling as a two-stage (cluster + within cluster stratified) if needed.
