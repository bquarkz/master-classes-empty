## AutoML using PyCaret Library

PyCaret is a low-code machine learning library in Python that automates the end-to-end machine learning workflow. It simplifies model building, comparison, tuning, and deployment into just a few lines of code, making it ideal for rapid experimentation and production-ready pipelines.

PyCaret enables users to quickly train and evaluate multiple machine learning algorithms without manually handling preprocessing, feature engineering, or hyperparameter tuning.


## Background

The primary objective of the analysis is to predict whether a customer will be classified as a “bad” credit risk (loan default or high-risk borrower) based on their financial liabilities, account behavior, employment stability, and personal demographics.

### 1. Import Required Library
We import pandas to load and manipulate the dataset.


```python
import pandas as pd
```

### 2. Load Dataset
Reads the customer master CSV file into a pandas DataFrame.


```python
df = pd.read_csv("Credit Scoring Data.csv")
df.head()
```

```python
df.info()
df.describe()
```

### 3. Convert Specified Columns to Factor (Categorical)

The following columns are business factor variables.
We convert them to categorical type so PyCaret treats them correctly.


```python
# Columns to convert to categorical
cols_to_factor = [
    "preloan", "veh", "house", "selfemp", "account", 
    "deposit", "branch", "ref", "age", "gender", 
    "ms", "child", "zone"
]

# Convert to categorical
for col in cols_to_factor:
    df[col] = df[col].astype("category")

# Define target
target = "bad"

# Drop ID columns
drop_cols = ["custid"]
df_model = df.drop(columns=drop_cols)

df_model.head()
```

### 4. Construct Setup function in PyCaret Environment
The `setup()` function initializes the entire machine learning pipeline.
It prepares data, defines preprocessing steps, splits data, and configures validation.
No models are trained at this stage. (data=) and (target=) are the minimal parameters that should be passed, without them PyCaret has no problem definition
- data = df (Pandas Dataframe)
- target = "bad" (Defines target variable, PyCaret should predict)
- train_size = 0.8  (Splits data into 80% train / 20% test, if not passed default train:test split is 70:30)
- session_id = 42 (random seed for reproducible results)
- normalize = True (scales numeric features)
- remove_multicollinearity = True (Removes highly correlated features)
- multicollinearity_threshold = 0.9 (features that have correlation above 0.9 are dropped)
  NOTE: For Multicollinearity, PyCaret uses Correlation values to eliminate features because of its efficiency in computation, and NOT Variance Inflation Factor (VIF) because it needs multiple regressions in the backend to obtain VIF values.
- verbose = False (controls logging noise)
- categorical_features = ["gender","branch",...] (forces PyCaret to treat integer coded columns as factors; used when data type conversion is not performed manually)
- fix_imbalance = True (applies SMOTE or similar techniques to handle Class Imbalance)
- fold_strategy = "kfold"/"timeseries" (controls cross validation style/critical for time dependent data)

```python
from pycaret.classification import *

clf_setup = setup(
    data=df_model,
    target=target,
    train_size=0.8,
    session_id=42,
    normalize=True,
    remove_multicollinearity=True,
    multicollinearity_threshold=0.9,
    verbose=False
)
```

### 5. Compare Multiple Models (AutoML)

Trains several classification algorithms using cross-validation
and ranks them based on AUC.

`n_select=1` returns the best performing model, but would still run all the algorithms. If n = 2 OR 3, it will run all the algorithms but return corresponding number of best models


```python
best_model = compare_models(
    n_select=1,
    sort="AUC"
)
```

### 6. View Best Model

Displays the top-performing model selected by AutoML.


```python
best_model
```

### 7. Hyperparameter Tuning

Optimizes the best model's parameters
to improve performance.


```python
tuned_model = tune_model(best_model)
```
