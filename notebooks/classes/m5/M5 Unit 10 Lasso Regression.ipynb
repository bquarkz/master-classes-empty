# Lasso Regression Demonstration In Python


## Background

Modern datasets (genomics, text, finance) often have:
- Thousands of predictors (p)
- Very small sample sizes (n)
- Strong multicollinearity

Ordinary Least Squares becomes unstable or infeasible. Regularization methods such as **LASSO** and **Elastic Net** provide:
- Shrinkage
- Variable selection
- Better generalization

## **Objective:** 
Demonstrate LASSO and Elastic Net for a *true high-dimensional regression problem* (p ≫ n) with a *continuous outcome*.


## Dataset Description

**Riboflavin Gene Expression Dataset**
- Samples (n): 71
- Predictors (p): 4,088 gene expression variables
- Target (Y): Riboflavin production rate (continuous)

This dataset is a canonical reference for LASSO / Elastic Net regression.

### Import Libraries

```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
```

### Load Data

```python
X, y = fetch_openml("riboflavin", version=1, return_X_y=True)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

### Save dataset as CSV for reuse

```python
df = X.copy()
df["target"] = y
df.to_csv("riboflavin_regression.csv", index=False)

print("CSV file saved: riboflavin_regression.csv")
```

### Split Data Into Train-Test 


```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

### Fit Lasso Regression

```python
lasso_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("lasso", Lasso(alpha=0.05, max_iter=10000))
])

lasso_pipeline.fit(X_train, y_train)
```

### Evaluate LASSO

```python
y_pred_lasso = lasso_pipeline.predict(X_test)

mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

lasso_coef = lasso_pipeline.named_steps["lasso"].coef_
nonzero_lasso = np.sum(lasso_coef != 0)

print("LASSO RESULTS")
print("---------------------------")
print(f"MSE: {mse_lasso:.4f}")
print(f"R² : {r2_lasso:.4f}")
print(f"Non-zero coefficients: {nonzero_lasso}")
```

### Fit Elastic Net Regression

```python
enet_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("enet", ElasticNet(alpha=0.05, l1_ratio=0.5, max_iter=10000))
])

enet_pipeline.fit(X_train, y_train)
```

###  Evaluate Elastic Net

```python
y_pred_enet = enet_pipeline.predict(X_test)

mse_enet = mean_squared_error(y_test, y_pred_enet)
r2_enet = r2_score(y_test, y_pred_enet)

enet_coef = enet_pipeline.named_steps["enet"].coef_
nonzero_enet = np.sum(enet_coef != 0)

print("ELASTIC NET RESULTS")
print("---------------------------")
print(f"MSE: {mse_enet:.4f}")
print(f"R² : {r2_enet:.4f}")
print(f"Non-zero coefficients: {nonzero_enet}")
```

### Model Comparison

```python
comparison = pd.DataFrame({
    "Model": ["LASSO", "Elastic Net"],
    "MSE": [mse_lasso, mse_enet],
    "R²": [r2_lasso, r2_enet],
    "Non-zero Coefficients": [nonzero_lasso, nonzero_enet]
})

comparison
```

### Conclusion
Elastic Net achieves a lower Mean Squared Error (MSE) than LASSO, indicating better predictive performance on the dataset.

### NOTE :
- LassoCV / ElasticNetCV are Python library functions (scikit-learn) that perform k-fold cross-validation to automatically select the optimal regularization parameter (α).

- For a grid of candidate α values, the model is fit on training folds and the mean squared error (MSE) is evaluated on validation folds.
- The optimal α is chosen as the value that minimizes the average cross-validated MSE, thereby balancing model fit and regularization.
