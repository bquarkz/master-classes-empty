## Matrix Operations

BACKGROUND:

You are given six observations on two variables X and Y.

X=[ 49.5, 54.6, 52.8,50.3,50.8,51.6]

Y=[ 36.9,56.7,53.8,43.1,49.4,45.4]

Here Y is the dependent variable and X is the independent variable.

## QUESTIONS

1. Perform Simple Linear Regression using ols function in Python and display table of regression coefficients.

```python
import pandas as pd
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings('ignore')

data = pd.DataFrame({
    'X': [49.5, 54.6, 52.8, 50.3, 50.8, 51.6],
    'Y': [36.9, 56.7, 53.8, 43.1, 49.4, 45.4]
})

model = smf.ols('Y ~ X', data=data).fit()
print(model.summary().tables[1])
```

2. Define design matrix X and target vector Y considering linear regression model with intercept


```python
import numpy as np

X_val = np.array([49.5, 54.6, 52.8, 50.3, 50.8, 51.6])
Y_vec = np.array([36.9, 56.7, 53.8, 43.1, 49.4, 45.4]).reshape(-1, 1)

# Design matrix with intercept (column of 1s)
X_mat = np.column_stack((np.ones(len(X_val)), X_val))

print("Design Matrix X:\n", X_mat)
print("\nTarget Vector Y:\n", Y_vec)
```

3. Obtain XT Y and (Xâ€™X)-1


```python
XT_Y = np.dot(X_mat.T, Y_vec)
XT_X_inv = np.linalg.inv(np.dot(X_mat.T, X_mat))

print("XT Y:\n", XT_Y)
print("\n(X'X)^-1:\n", XT_X_inv)
```

4. Obtain least square estimates of two model parameters


```python
beta_hat = np.dot(XT_X_inv, XT_Y)
print("Least Square Estimates (Intercept, Slope):\n", beta_hat)
```

5. Compare the answers with ols function output

```python
print("OLS Function Estimates:")
print(model.params)
print("\nMatrix Calculations Estimates:")
print(f"Intercept: {beta_hat[0][0]}")
print(f"Slope:     {beta_hat[1][0]}")
```
