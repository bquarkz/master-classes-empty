1. At a large hospital, 30% of patients visit for a routine check-up (Event A), and 20% visit for vaccination (Event B). Assuming independence, find P(A ∩ B).

```python
p_a = 0.30
p_b = 0.20
p_a_AND_b = p_a * p_b
print(f"P(A ∩ B) = {p_a_AND_b}")
```

2. In a supermarket, 40% buy fruits (F). Among them, 25% buy yogurt (Y). Find P(Y | F).

```python
p_f = 0.40
p_y_GIVEN_f = 0.25
print(f"P(Y | F) = {p_y_GIVEN_f}")
```

3. 12% of adults have high BP (H). Among them, 70% are aware (A). Find P(A ∩ H).

```python
p_h = 0.12
p_a_GIVEN_h = 0.70
p_a_AND_h = p_a_GIVEN_h * p_h
print(f"P(A ∩ H) = {p_a_AND_h}")
```

4. In a shopping app, 45% browse electronics (E), 30% browse home appliances (H), and 12% browse both. Find P(E ∪ H).

```python
p_e = 0.45
p_h = 0.30
p_e_AND_h = 0.12
p_e_OR_h = p_e + p_h - p_e_AND_h
print(f"P(E ∪ H) = {p_e_OR_h}")
```

5. In an insurance company, 25% have car policy (C), 40% have health policy (H), and 18% have both. Find P(C ∪ H).

```python
p_c = 0.25
p_h = 0.40
p_c_AND_h = 0.18
p_c_OR_h = p_c + p_h - p_c_AND_h
print(f"P(C ∪ H) = {p_c_OR_h}")
```

6. Three fair coins are tossed simultaneously. You are told that at least two of the coins show Heads. What is the probability that all three coins show Heads, given this information?

```python
# Sample space S = {HHH, HHT, HTH, THH, HTT, THT, TTH, TTT}
# Event A (at least two heads) = {HHH, HHT, HTH, THH} -> 4 outcomes
# Event B (all three heads) = {HHH} -> 1 outcome
# P(B | A) = P(B ∩ A) / P(A) = 1/4
print(f"P(All Heads | At least two Heads) = {1 / 4}")
```
